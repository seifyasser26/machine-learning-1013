```
def computeCost(X, Y, theta, lmda):
    m = Y.size  
    J = 0
    h=np.dot(X,theta)
    J=(1/(2*m))*(sum((h-Y)**2))
    k=((lmda/(2*m))*sum((theta)**2))
    J=J+k
    return J
```

```
 def  featureNormalize(X):
    
      X_norm = X.copy()

      mu = np.mean(X)
      X_norm = X - mu

      sigma = np.std(X_norm)
      X_norm = X_norm / sigma
   
      return X_norm
```
